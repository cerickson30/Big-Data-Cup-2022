```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
nwhl_pbp_dat = pd.read_csv('../hackathon_nwhl.csv')

nwhl_pbp_dat.columns = [colname.lower().replace(' ', '_') for colname in nwhl_pbp_dat.columns]

nwhl_pbp_dat['game_date'] = pd.to_datetime(nwhl_pbp_dat.game_date, format='%Y-%m-%d').dt.date

nwhl_pbp_dat['clock'] = nwhl_pbp_dat.clock.apply(lambda x: 60 * int(x.split(':')[0]) + int(x.split(':')[1]))
nwhl_pbp_dat.rename({'clock':'clock_seconds', 'x_coordinate':'x_coord', 'x_coordinate_2':'x_coord_2', 'y_coordinate':'y_coord', 'y_coordinate_2':'y_coord_2'}, axis=1, inplace=True)
```

```python
nwhl_pbp_dat.head()
```

```python
oly18_pbp_dat = pd.read_csv('../hackathon_womens.csv')

oly18_pbp_dat.columns = [colname.lower().replace(' ', '_') for colname in oly18_pbp_dat.columns]

oly18_pbp_dat['game_date'] = pd.to_datetime(oly18_pbp_dat.game_date, format='%Y-%m-%d').dt.date

oly18_pbp_dat['clock'] = oly18_pbp_dat.clock.apply(lambda x: 60 * int(x.split(':')[0]) + int(x.split(':')[1]))
oly18_pbp_dat.rename({'clock':'clock_seconds', 'x_coordinate':'x_coord', 'x_coordinate_2':'x_coord_2', 'y_coordinate':'y_coord', 'y_coordinate_2':'y_coord_2'}, axis=1, inplace=True)
```

```python
oly18_pbp_dat.head()
```

```python
oly22_pbp_dat = pd.read_csv('../pxp_womens_oly_2022_v2.csv')
oly22_pbp_dat['game_date'] = pd.to_datetime(oly22_pbp_dat.game_date, format='%d/%m/%Y').dt.date
oly22_pbp_dat.drop('season_year', axis=1, inplace=True)
```
